```python
import xarray as xr
import numpy as np
```

```python
filename = 'moc_transports.nc'
ds = xr.open_dataset(filename)
```

```python
amoc = ds.moc_mar_hc10
```

```python
amoc_ann = amoc.groupby('time.year').mean()
```

```python
amoc_ann.plot()
```

```python
amoc_ann.year.values
```

```python
np.savez('amoc_ann.npz', amoc_ann=amoc_ann.values[:-1], time=amoc_ann.year.values[:-1])
```
