```python
import numpy as np
import pandas as pd
```

```python
siarea = {}
year = {}
for key in ['N_03', 'N_09', 'S_03', 'S_09']:
    df = pd.read_csv(key+'_extent_v3.0.csv')
    year[key] = df.iloc[:,0].to_numpy()
    siarea[key] = df.iloc[:,5].to_numpy()
```

```python
import matplotlib.pyplot as plt
```

```python
for key in ['N_03', 'N_09', 'S_03', 'S_09']:
    plt.plot(year[key], siarea[key])
```

```python
np.savez('siarea_MarSep.npz', siarea_n03=siarea['N_03'], siarea_n09=siarea['N_09'],
         siarea_s03=siarea['S_03'], siarea_s09=siarea['S_09'],
         time=year['N_03'])
```
