```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
# figure root
fig_root = os.environ['HOME']+'/work/E3SMv1_Future_Projection/figures'
save_fig = True
```

```python
e3sm_sst_data = np.load('e3sm_sst.npz')
esgf_sst_data = np.load('esgf_sst.npz')
```

```python
esgf_colors = np.load('cmip6_colors.npz', allow_pickle=True)['esgf_colors'].flat[0]
```

```python
fig = plt.figure(figsize=(6, 6))

esgf_time = esgf_sst_data['time']
e3sm_time = e3sm_sst_data['time']
models = []
for i, s in enumerate(esgf_sst_data['label']):
    esgf_sst = esgf_sst_data['data'][i,:]
    lb = s.split('_')[0]
    if lb not in models:
        models.append(lb)
        plt.plot(esgf_time, esgf_sst, color=esgf_colors[lb], label=lb)
    else:
        plt.plot(esgf_time, esgf_sst, color=esgf_colors[lb])
    
for i, s in enumerate(e3sm_sst_data['label']):
    e3sm_sst = e3sm_sst_data['data'][i,:]
    if s.startswith('E3SM P'):
        lb = 'E3SM'
        color = 'k'
    elif s.startswith('E3SM GHG'):
        lb = 'E3SM'
        color = 'gray'
    if s.endswith('1'):
        plt.plot(e3sm_time, e3sm_sst, color=color, label=lb)
    else:
        plt.plot(e3sm_time, e3sm_sst, color=color)

plt.xlim([2015, 2100])
plt.ylabel('SST ($^\circ$C)')
plt.legend(loc='lower center',  bbox_to_anchor=(0.5, -0.7), ncol=3)

# plt.tight_layout()
plt.subplots_adjust(bottom=0.4)
# save figure
if save_fig:
    figname = fig_root+'/CMIP6_ScenarioMIP_SST.png'
    fig.savefig(figname, dpi = 300)
```
